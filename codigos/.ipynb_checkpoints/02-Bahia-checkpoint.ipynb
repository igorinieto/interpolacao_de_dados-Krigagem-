# Trabalhando com as informações do estado da Bahia

## 1. Importando bibliotecas

```python
import pandas as pd 
from IPython.display import Image
import funcoes_auxiliares as fa
```

## 2. Abrindo csv do PronaSolos após tratamento

```python
# Mostrar todas as colunas
pd.set_option('display.max_columns', None) 
# Carregando o csv
df = pd.read_csv('../dados/pronasolos.csv')
df
```

## 3. Selecionando informações do estado da Bahia

```python
estado = 'BA'
df_estado = df[df['uf'] == estado]
print(f'Quantidade de cidades no estado de {estado}: ',len(df_estado['municipio'].unique()))
df_estado
```

**Lista das características que iremos observar os variogramas**

```python
lista_caracteristicas = ['cascalho', 'terra_fina','areia_gros', 'areia_fina', 'areia_tota', 'silte', 'argila',
       'argila_dis', 'grau_flocu', 'relacao_si','ph_h2o', 'ph_kcl', 'complexo_s', 'complexo_1', 'complexo_2', 'complexo_3',
       'valor_s', 'aluminio_t', 'hidrogenio', 'valor_t', 'valor_v', 'saturacao_', 'nitrogenio']
```

## 4. Criação de dataframe de cada característica da lista de características e plot dos dados

```python
dic_df = {}
for i, caracteristica in enumerate(lista_caracteristicas):
    df = fa.df_caracteristica(caracteristica, df_estado)
    dic_df[caracteristica] = df
```

**Lista das características escolhidas, segundo a distribuição dos dados, para realizar os variogramas.**

```python
lista_caracteristicas_escolhidas = ['areia_gros','areia_fina', 'areia_tota', 'silte','argila_dis', 'grau_flocu', 'ph_kcl']
```

## 5. Exportando lista de DataFrames das características selecionadas para krigagem no R 

```python
ba = fa.export_caracteristicas_escolhidas(lista_caracteristicas_escolhidas, zona = '24', estado = 'ba', dic_df = dic_df)
```

## 6. Interpolação  no R 
**Utilizando o script "interpolacao_krigagem.R" disponibilizado na pasta códigos.**
